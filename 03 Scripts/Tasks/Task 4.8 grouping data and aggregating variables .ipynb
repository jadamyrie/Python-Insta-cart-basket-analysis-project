```python
# group the df dataframe by the 'department_id' column, then calculate the mean, min and max of the 'order_number' column
print('The mean, min and max of order_number grouped by the department_id column in df:')
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
# Create a new column, 'max_order' to place the results of the maximum orders for each user.
```

```python
# create a new column 'max_order' containing the maximum order number for each user_id, calculated using the groupby() and transform() functions
df_ords_merged['max_order'] = df_ords_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# check the output of df
print('first 10 rows of dataframe')
df_ords_merged.head(10)
```

```python
Create a flag that assigns a loyalty label to a user ID based on its corresponding max order value.
```

```python
# create the 'loyalty_flag' column based on each user's maximum order value
df_ords_merged.loc[df_ords_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_merged.loc[(df_ords_merged['max_order'] <= 40) & (df_ords_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_merged.loc[df_ords_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# frequency in the 'loyalty_flag' column
print('The frequency in the loyalty_flag column:')
df_ords_merged['loyalty_flag'].value_counts(dropna = False)
```

```python
# check the accuracy of the 'loyalty_flag' column with output from columns 'user_id', 'max_order', and 'loyalty_flag' column
print('Sample output from columns user_id, max_order and loyalty_flag columns:')
df_ords_merged[['user_id', 'max_order', 'loyalty_flag']].head(60)
```

# Do the prices of products purchased by loyal customers differ from those purchased by regular or new customers?

```python
# basic statistics of df_ords_merged by the 'loyalty_flag' column, then calculate the mean, min and max of the 'prices' column
print('The mean, min and max of prices grouped by the loyalty_flag column in df_ords_merged:')
df_ords_merged.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
# creating spending_flag column using transform()
df_ords_merged['average_price'] = df_ords_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# first rows
df_ords_merged.head()
```

```python
 Create a spending flag for each user based on the average price across all their orders
```

```python
# create the 'spending_flag' column based on each user's average price
df_ords_merged.loc[df_ords_merged['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_ords_merged.loc[df_ords_merged['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# count the frequency in the 'spending_flag' column
print('The frequency in the spending_flag column:')
df_ords_merged['spending_flag'].value_counts(dropna = False)
```

```python
# check the accuracy of the 'spending_flag' column with output from columns 'user_id', 'average_price', and 'loyalty_flag' column
print('first 30 output from columns user_id, average_price and spending_flag columns:')
df_ords_merged[['user_id', 'average_price', 'spending_flag']].head(30)
```

```python
# create a new column 'order_freq' containing the median in days_since_prior_order for each user_id, calculated using the groupby() and transform() functions
df_ords_merged['order_freq'] = df_ords_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# check the output of 'user_id' and 'order_freq'
print('first 10 user_id and order_freq output:')
df_ords_merged[['user_id', 'order_freq']].head(10)
```

```python
# create the 'order_freq_flag' column based on each user's median in days_since_prior_order
df_ords_merged.loc[df_ords_merged['order_freq'] > 20, 'order_freq_flag'] = 'Non-frequent customer'
```

```python
df_ords_merged.loc[(df_ords_merged['order_freq'] > 10) & (df_ords_merged['order_freq'] <= 20), 'order_freq_flag'] = 'Regular customer'
```

```python
df_ords_merged.loc[df_ords_merged['order_freq'] <= 10, 'order_freq_flag'] = 'Frequent customer'
```

```python
# frequency of the 'order_freq_flag' column
print('The frequency of the order_freq_flag column:')
df_ords_merged['order_freq_flag'].value_counts(dropna = False)
```
